```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('상권변화지표.csv', thousands=',', encoding='euc-kr').drop(['Unnamed: 0'], axis=1)
df
```

```python
import matplotlib.pyplot as plt
```

```python
columns = ['개업률', '폐업률', '신생 1  생존률', '신생 3  생존률', '신생 5  생존률',
       '1  생존률', '3  생존률', '5  생존률', '주거 인구', '직장 인구', '임대료', '전체 점포수',
       '프랜차이즈 점포수', '일반 점포수', '최근 10년  평균영업기간', '최근 30년  평균영업기간']
df1 = pd.DataFrame(df, columns=columns)
df1['target'] = df['상권변화지표']
df1
```

```python
from sklearn.preprocessing import StandardScaler

df_scaled = StandardScaler().fit_transform(df1.iloc[:, :-1])
df_scaled
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=4)
pca.fit(df_scaled)
df_pca = pca.transform(df_scaled)
```

```python
pca_columns = ['pca_component_1', 'pca_component_2', 'pca_component_3', 'pca_component_4']
df_pca2 = pd.DataFrame(df_pca, columns=pca_columns)
df_pca2['target'] = df['상권변화지표']
df_pca2
```

```python
markers = ['^', 's', 'o', '*']
```

```python
df_pca2['target'].unique()[0]
```

```python
markers = ['^', 's', 'o', '*']

for i, marker in zip(df_pca2['target'].unique(), markers):
    x_data = df_pca2[df_pca2['target']==i]['pca_component_1']
    y_data = df_pca2[df_pca2['target']==i]['pca_component_2']
    plt.scatter(x_data, y_data, marker=marker, label=df_pca2['target'].unique())

plt.legend()
plt.xlabel('pca_component_1')
plt.ylabel('pca_component_2')
plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier
```

```python
data = df[columns]
target = df['상권변화지표']
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.2)
```

```python
x_train.shape, x_test.shape
```

```python
rf = RandomForestClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(rf, x_train, y_train, return_train_score=True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
rf.fit(x_train, y_train)
```

```python
rf.predict([np.array(x_train)[0]])
```
