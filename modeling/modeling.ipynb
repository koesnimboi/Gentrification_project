```python
df = pd.read_csv('상권변화지표.csv', thousands=',',encoding='euc-kr').drop(['Unnamed: 0'], axis=1)
df
```

```python
a = df['상권변화지표']
df['프랜차이즈 점포 비율'] = df['프랜차이즈 점포수'] / df['전체 점포수']
df['직장 인구 대비 주거 인구'] = df['직장 인구'] / df['주거 인구']
df = df.drop(['전체 점포수', '프랜차이즈 점포수', '일반 점포수', '상권변화지표', '직장 인구', '주거 인구'], axis=1)
df['상권변화지표'] = a
df
```

```python
df['행정구역']
```

```python
columns = ['개업률', '폐업률', '신생 1  생존률', '신생 3  생존률', '신생 5  생존률',
       '1  생존률', '3  생존률', '5  생존률', '임대료', '최근 10년  평균영업기간', '최근 30년  평균영업기간',
       '프랜차이즈 점포 비율', '직장 인구 대비 주거 인구']
df1 = pd.DataFrame(df, columns=columns)
df1['target'] = df['상권변화지표']
df1
```

```python
df2 = df1.drop(['신생 1  생존률', '신생 3  생존률', '신생 5  생존률', '1  생존률', '3  생존률', '최근 30년  평균영업기간', '임대료'], axis=1)
```

#### 정체 -> 0, 다이나믹 -> 1, '상권축소' -> 2, '상권확장' -> 3 로 치환

```python
df2['target'] = df2['target'].map({'정체': 0, '다이나믹': 1,'상권축소' : 2, '상권확장' : 3},na_action=None)
```

```python
df2
```

```python
X = df2.iloc[:, :-1]
Y = df2.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X_scaled = StandardScaler().fit_transform(X)

data_train = X_scaled[:450]
target_train = Y[:450]
data_test = X_scaled[450:]
target_test = Y[450:]

x_train, x_val, y_train, y_val = train_test_split(data_train, target_train, test_size=0.2)

x_train.shape, x_val.shape, y_train.shape, y_val.shape
```

## Model selection

```python
# importing library

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression 
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import RidgeClassifier
from sklearn.ensemble import GradientBoostingClassifier as GBM
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.svm import SVC
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

from sklearn import preprocessing
from sklearn.model_selection import cross_val_score , ShuffleSplit
from sklearn.model_selection import cross_validate
from sklearn import metrics
from sklearn.metrics import roc_curve , roc_auc_score , f1_score , classification_report, accuracy_score
```

```python
# 랜덤 포레스트 accuracy 계산
model_rf = RandomForestClassifier(n_estimators=50, n_jobs=4, random_state=42).fit(x_train, y_train)
pred_rf = model_rf.predict(x_val)
accuracy_rf = accuracy_score(y_val, pred_rf)

# 로지스틱 회귀 accuracy 계산
model_lr = LogisticRegression(multi_class='multinomial', solver='lbfgs', C=10, random_state=42).fit(x_train, y_train)
pred_lr = model_lr.predict(x_val)
accuracy_lr = accuracy_score(y_val, pred_lr)

# 결정 트리 accuracy 계산
model_dt = DecisionTreeClassifier(random_state = 42).fit(x_train, y_train)
pred_dt = model_dt.predict(x_val)
accuracy_dt = accuracy_score(y_val, pred_dt)

# 릿지 classifier accuracy 계산
model_clf = RidgeClassifier(solver='auto', random_state = 42).fit(x_train, y_train)
pred_clf = model_clf.predict(x_val)
accuracy_clf = accuracy_score(y_val, pred_clf)

# 히스토그램 기반 그래디언트 부스팅 accuracy 계산
model_hgb = HistGradientBoostingClassifier(learning_rate = 0.1, max_iter = 1000, random_state=42).fit(x_train, y_train)
pred_hgb = model_hgb.predict(x_val)
accuracy_hgb = accuracy_score(y_val, pred_hgb)
```

```python
idxs = ['RandomForestClassifier', 'LogisticRegression', 'DecisionTreeClassifier', 'RidgeClassifier', 'HistGradientBoostingClassifier']
scores = [accuracy_rf, accuracy_lr, accuracy_dt, accuracy_clf, accuracy_hgb]
data = list(zip(idxs, scores))

output = pd.DataFrame(data, columns = ['model', 'score'])

plt.figure(figsize=(5,5))
sns.set(font_scale=1.5)
sns.barplot(x = 'score', y = 'model', data = output, orient='h')
plt.xlim(0,1.0)
plt.title('Model Score Comparison')
```
